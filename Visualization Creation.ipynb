```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

```

```python
oregon_df = pd.read_csv("Data/oregon_data.csv")

oregon_df.head()
```

```python
oregon_aggassault = oregon_df["Aggravated Assault"]
year = oregon_df["Year"]

plt.plot(year, oregon_aggassault)
plt.xlabel("Year")
plt.ylabel("Number of Aggravated Assault Cases")
plt.title("Oregon-Aggravated Assault Cases 2011-2020")

plt.show()
```

```python
oregon_drug_abuse = oregon_df["Drug Abuse Violations"]
year = oregon_df["Year"]

plt.plot(year, oregon_drug_abuse)
plt.xlabel("Year")
plt.ylabel("Number of Drug Abuse Violations")
plt.title("Oregon-Drug Abuse Violations 2011-2020")

plt.show()
```

```python
oregon_weapons = oregon_df["Weapons: Carrying, Possessing, Etc."]
year = oregon_df["Year"]

plt.plot(year, oregon_weapons)
plt.xlabel("Year")
plt.ylabel("Number of Weapons Violations")
plt.title("Oregon-Weapons Violations 2011-2020")

plt.show()
```

```python
oregon_aggassault = oregon_df["Aggravated Assault"]
oregon_weapons = oregon_df["Weapons: Carrying, Possessing, Etc."]
oregon_drug_abuse = oregon_df["Drug Abuse Violations"]
year = oregon_df["Year"]

plt.plot(year, oregon_aggassault, color="red", marker=".")
plt.plot(year, oregon_weapons, marker=".")
plt.plot(year, oregon_drug_abuse, color ="green", marker=".")

#plt.xlabel("Year")
#plt.ylabel("Number of Aggravated Assault Cases")
#plt.title("Oregon-Aggravated Assault Cases 2011-2020")

plt.show()
```
